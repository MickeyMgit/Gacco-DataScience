```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random

an = 50000000
sp = []
sp = np.random.geometric(0.075, an)
sp *= 10000
rn = int(1 + np.log2(an))
plt.hist(sp, rn)
plt.title("1.Population Distribution")
plt.show()
print("1.母集団平均値 = ", np.mean(sp), ", 階級数 = ", rn)
print("")

random.seed(1234)
sn = 9000
rsp = []
rsp = random.sample(list(sp), sn)
rn = int(1 + np.log2(sn))
plt.hist(rsp, rn)
plt.title("2.Sample Distribution")
plt.show()
print("2.標本平均値 = ", np.mean(rsp), ", 階級数 = ", rn)
print("")

rep = 500
ext = 1000
rspl = []
for i in range(rep):
  rrsp = []
  rrsp = random.sample(rsp, ext)
  rspl.append(rrsp)
mrspl = []
for i in range(rep):
  mrsp = np.mean(rspl[i])
  mrspl.append(mrsp)
sn = ext
rn = int(1 + np.log2(sn))
plt.hist(mrspl, rn)
plt.title("3(a).Sample Mean Distribution")
plt.show()
print("繰り返し数 = ", rep, ", 抽出数 = ", ext, ", 階級数 = ", rn)
print("")

bnum = int(max(mrspl) - min(mrspl)) // rn
bucket = {}
for i in range(rn):
  dic = {int(min(mrspl) + bnum * i):0}
  bucket.update(dic)
for i in range(len(mrspl)):
  bkey = next(iter(bucket))
  for key in bucket:
    if (mrspl[i] < key):
      bucket[bkey] += 1
      break
    bkey = key
exb = 0
for key in bucket:
  exb += (key / len(mrspl)) * bucket[key]
print("3(a).期待値 = ", exb)
print("")
print("")

extnum = 20
sn = 9000
erspl = []
for i in range(extnum):
  ersp = []
  ersp = random.sample(list(sp), sn)
  erspl.append(ersp)
emrspl = []
for i in range(extnum):
  mrsp = np.mean(erspl[i])
  emrspl.append(mrsp)
rn = int(1 + np.log2(sn))
plt.hist(emrspl, rn)
plt.title("3(b).Sample Mean Distribution")
plt.show()
print("繰り返し数 = ", extnum, ", 抽出数 = ", sn, ", 階級数 = ", rn)
print("")

ebnum = int(max(emrspl) - min(emrspl)) // extnum
ebucket = {}
for i in range(extnum):
  dic = {int(min(emrspl) + ebnum * i):0}
  ebucket.update(dic)
for i in range(len(emrspl)):
  bkey = next(iter(ebucket))
  for key in ebucket:
    if (emrspl[i] < key):
      ebucket[bkey] += 1
      break
    bkey = key
exb = 0
for key in ebucket:
  exb += (key / len(emrspl)) * ebucket[key]
print("3(b).期待値 = ", exb)
print("")
```
